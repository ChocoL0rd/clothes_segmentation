```python
import json

import os
import shutil
import cv2
from PIL import Image
```

```python
bad_imgs = [
    "5a425372e44464232f28d93a55a0fd37",
    "a360753cbc5390f0a1fbac5c7c74e35b",
    "aa38dbb2761750a5a1f31be6eb24cff3",
    "bbec629ac9252dfb2fdbdaec23f16db2",
    "e83dd4eb91673bba788df17c31460de0",
    "415fed695d1b17b756da2caa6972d6ec",
    "12ed3e2741db79d0c842f26d9a85fb5d",
    "4e560d55682ae59ac63e387d873f4876",
    
    "2deabee92e31dbc0e11e475d67791228",
    "9c33cd4bc519c5d7c23cd918d8fad65a",
    
    "8267c9506a6886abaa40a6515d52b3fe"
]
print(f"length: {len(bad_imgs)}")

# fix dataset 
with open("datasets.json", "r") as f:
    old_dict = json.load(f)

old_train_size = sum([len(i) for i in old_dict['datasets']['train'].values()])
old_test_size = sum([len(i) for i in old_dict['datasets']['test'].values()])
    
print(f"train_size: {old_train_size}")    
print(f"test_size: {old_test_size}")

new_dict = old_dict.copy()

ds_names = ["train", "test"]

for ds_name in ds_names:
    for dir_name, value in new_dict["datasets"][ds_name].items():
        for img_name in bad_imgs:
            if img_name in value:
                value.remove(img_name)

        new_dict["datasets"][ds_name][dir_name] = value

new_train_size = sum([len(i) for i in new_dict['datasets']['train'].values()])
new_test_size = sum([len(i) for i in new_dict['datasets']['test'].values()])

print(f"new train_size: {new_train_size}")    
print(f"new test_size: {new_test_size}")  


print(f"difference: {new_train_size + new_test_size - old_train_size - old_test_size}")
```

```python
with open("fix1_datasets.json", "w") as f:
    json.dump(new_dict, f, indent=4)
```
